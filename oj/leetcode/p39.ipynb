```python
class Solution:
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        
        rv = []
        candidates = sorted(candidates, reverse=True)
        
        
        def dfs(s, idx, current_sum):
            if idx >= len(candidates):
                return
            
            # use c, 
            c = candidates[idx]
            new_sum = current_sum
            next_s = s
            while new_sum < target:
                next_s = next_s + [c]
                new_sum = new_sum + c
                if new_sum > target:
                    break
                elif new_sum == target:
                    rv.append(next_s)
                elif new_sum < target:
                    dfs(next_s, idx + 1, new_sum)
                    
            # not use c
            dfs(s, idx + 1, current_sum)
                
        dfs([], 0, 0)
        
        return rv
        
        
        
s = Solution()


def check(a, b):
    if len(a) != len(b):
        return False
    a = sorted([sorted(i) for i in a])
    b = sorted([sorted(i) for i in b])
    for idx in range(len(a)):
        if a[idx] != b[idx]:
            return False
    return True


for idx, (c, t, e) in enumerate([
    ([9, 1], 1, [[1]]),
    ([2, 3, 6, 7], 7,
     [[7], [2, 2, 3]]),
    ([2, 3, 5], 8,
     [[2, 2, 2, 2],
      [2, 3, 3],
      [3, 5]]),
]):
    rv = s.combinationSum(c, t)
    if check(rv, e):
        print(f'{idx} passed')
    else:
        print(f'{idx} failed: {rv}')
```
