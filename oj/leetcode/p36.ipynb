```python
def print_board(board):
    print('\n'.join(' '.join(l) for l in board))
    print()


class Solution:
    def solveSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: void Do not return anything, modify board in-place instead.
        """
        
        def valid_x(d, x):
            for y in range(9):
                if board[x][y] == d:
                    return False
            return True
        
        def valid_y(d, y):
            for x in range(9):
                if board[x][y] == d:
                    return False
            return True
        
        def valid_s(d, x, y):
            xb, yb = x // 3 * 3, y // 3 * 3
            for dx in range(3):
                for dy in range(3):
                    if board[xb + dx][yb + dy] == d:
                        return False
            return True
        
        def next_xy(x, y):
            if x + 1 < 9:
                return x + 1, y
            return 0, y + 1
        
        def dfs(x, y):
            if board[x][y] != '.':
                return dfs(*next_xy(x, y))
                
            nx, ny = next_xy(x, y)
            for d in '123456789':
                if not valid_x(d, x):
                    continue
                if not valid_y(d, y):
                    continue
                if not valid_s(d, x, y):
                    continue
                board[x][y] = d
                if x == 8 and y == 8:
                    return True
                if dfs(nx, ny):
                    return True
                board[x][y] = '.'
            return False
        
        dfs(0, 0)
        
        
s = Solution()

b = [
    ["5","3",".",".","7",".",".",".","."],
    ["6",".",".","1","9","5",".",".","."],
    [".","9","8",".",".",".",".","6","."],
    ["8",".",".",".","6",".",".",".","3"],
    ["4",".",".","8",".","3",".",".","1"],
    ["7",".",".",".","2",".",".",".","6"],
    [".","6",".",".",".",".","2","8","."],
    [".",".",".","4","1","9",".",".","5"],
    [".",".",".",".","8",".",".","7","9"],
]

s.solveSudoku(b)

print_board(b)

assert b == [
    ["5","3","4","6","7","8","9","1","2"],
    ["6","7","2","1","9","5","3","4","8"],
    ["1","9","8","3","4","2","5","6","7"],
    ["8","5","9","7","6","1","4","2","3"],
    ["4","2","6","8","5","3","7","9","1"],
    ["7","1","3","9","2","4","8","5","6"],
    ["9","6","1","5","3","7","2","8","4"],
    ["2","8","7","4","1","9","6","3","5"],
    ["3","4","5","2","8","6","1","7","9"],
]
```
